```python
#importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations=1000
nx=np.random.uniform(low=-10,high=10,size=(observations,1))
nz=np.random.uniform(low=-10,high=10,size=(observations,1))
inputs=np.column_stack((nx,nz))
inputs.shape
```

```python
noise=np.random.uniform(-1,1,(observations,1))
target=nx*2-nz*3+5+noise
target.shape
```

```python
target=target.reshape(observations,)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot(nx,nz,target)
ax.set_xlabel("NX")
ax.set_ylabel("NZ")
ax.set_zlabel("Targets")
ax.view_init(azim=10)
plt.show()
target=target.reshape(observations,1)
```

```python
init_range=0.1
weights=np.random.uniform(low=-init_range,high=init_range,size=(2,1))
biases=np.random.uniform(low=-init_range,high=init_range,size=(1))
weights
biases
```

```python
learning_rate=0.02
```

```python
for i in range(100):
    y=np.dot(inputs,weights)+biases
    deltas=y-target
    loss=np.sum(deltas**2)/2/observations
    print(loss)
    deltas_scaled=deltas/observations
    weights=weights-learning_rate * np.dot(inputs.T,deltas_scaled)
    biases=biases-learning_rate*np.sum(deltas_scaled)
```

```python
print(weights,biases)
```
