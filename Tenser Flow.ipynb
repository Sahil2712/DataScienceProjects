```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observations=1000
nx=np.random.uniform(low=-10,high=10,size=(observations,1))
nz=np.random.uniform(low=-10,high=10,size=(observations,1))
generated_inputs=np.column_stack((nx,nz))
noise=np.random.uniform(-1,1,(observations,1))
generated_targets=2*nx-3*nz+5+noise
np.savez('Tf_intro',inputs=generated_inputs,targets=generated_targets)
```

```python
#Loading data and storing in training_data
training_data=np.load('Tf_intro.npz')
```

```python
#Displaying data
training_data
```

```python
input_size=2
output_size=1
model=tf.keras.Sequential([tf.keras.layers.Dense(output_size,
                           kernel_initializer=tf.random_uniform_initializer(-0.1,0.1),
                       # bias_initialzer=tf.random_uniform_initializer(-0.1,0.1)
                                                bias_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1))])
customize_data=tf.keras.optimizers.SGD(learning_rate=0.02)
model.compile(optimizer=customize_data,loss='mean_squared_error')
model.fit(training_data['inputs'],training_data['targets'],epochs=100,verbose=2)
```

```python
model.layers[0].get_weights()
```

```python
weights=model.layers[0].get_weights()[0]
weights
```

```python
biases=model.layers[0].get_weights()[1]
biases
```

```python
#To predict the values
model.predict_on_batch(training_data['inputs'])
```

```python
training_data['targets'].round(1)
```

```python
#plotting graph
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])),np.squeeze(training_data['targets']))
plt.xlabel("OUTPUTS")
plt.ylabel("Targets")
plt.show()
```
