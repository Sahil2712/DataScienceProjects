# Loan Prediction

```python
#importing the standard libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#Reading trained Data
train_data=pd.read_csv('train_ctrUa4K.csv')
```

```python
#Reading test Data
test_data=pd.read_csv('test_lAUu6dG.csv')
```

```python
#Displaying first 5 rows of trained data
train_data.head()
```

```python
#Displaying first 5 rows of test data
test_data.head()
```

```python
#printing shape of train and test data
print("Shape of train and Test data respectively:",train_data.shape,test_data.shape)
```

```python
#Transforming Loan Status to either 0 or 1
le=LabelEncoder()
train_data.Loan_Status=le.fit_transform(train_data.Loan_Status)
train_data.Loan_Status
```

```python
sum(train_data.Loan_Status)
```

```python
#Dropping Null values
train_data.dropna(how='any',inplace=True)
train_data.shape
```

```python
#Splitiing train data in train and test
train,test=train_test_split(train_data,test_size=0.2,random_state=0)
```

```python
#Displaying train data after splitting train dataset
train.head()
```

```python
#displaying test data after splitting train data
test
```

```python
#Dropping Loan ID and Loan Status and storing  the dataset in train_x
train_x=train.drop(columns=['Loan_ID','Loan_Status'],axis=1)
#storing Loan Status in train_y
train_y=train['Loan_Status']
```

```python
#Displaying train_x
train_x
```

```python
#Dropping Loan ID and Loan Status and storing  the dataset in test_x
test_x=test.drop(columns=['Loan_ID','Loan_Status'],axis=1)
#storing Loan Status in test_y
test_y=test['Loan_Status']
```

```python
#Displaying test_x
test_x.head()
```

```python
#Using Dummies seperating the files whichever is present in the application will pe displayed as 1 and if its not present it will be displayed as 0
train_x=pd.get_dummies(train_x)
#Displaying train_x dataset after using dummies
train_x
```

```python
#Using Dummies seperating the files whichever is present in the application will pe displayed as 1 and if its not present it will be displayed as 0
test_x=pd.get_dummies(test_x)
#Displaying train_x dataset after using dummies
test_x
```

```python
#Printing shape of test_x and train_x
print('',test_x.shape,train_x.shape)
```

```python
#Applying Logistic regressinon
lr=LogisticRegression()
#Fitting train_x and train__y in regression
lr.fit(train_x,train_y)
```

```python
#Predicting the test_x data set
predict=lr.predict(test_x)
```

```python
#Displaying predicted values
predict
```

```python
#Checking accuracy of the regression
accuracy_score(test_y,predict)
```

```python
#Checking intercept of regression
lr.intercept_
```
