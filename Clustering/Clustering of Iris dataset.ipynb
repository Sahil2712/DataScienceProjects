```python
#imprting Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
#Reading data
data=pd.read_csv('iris-dataset.csv')
```

```python
#displaying data
data.head()
```

```python
#plotting graph
plt.scatter(data['sepal_length'],data['sepal_width'])
plt.xlabel('Length of sepal')
plt.ylabel('Width of Sepal')
plt.show()
```

```python
#Applying clusters
x=data.copy()
kmeans=KMeans(2)
kmeans.fit(x)
```

```python
#Predicting clusters
cluster=kmeans.fit_predict(x)
cluster
```

```python
#Creating an cluster_predict column and storing the predicted cluster
data['Cluster_predict']=cluster
data
```

```python
#Plotting graph
plt.scatter(data['sepal_length'],data['sepal_width'],c=data['Cluster_predict'],cmap='rainbow')
```

```python
#Preprocessing data
from sklearn import preprocessing
x_scaled=preprocessing.scale(data)
x_scaled
```

```python

    kmeans=KMeans(2)
    kmeans.fit(x_scaled)
    clusters=data.copy()
    clusters['cluster_new_predict']=kmeans.fit_predict(x_scaled)
```

```python
#Displaying Clusters
clusters
```

```python
#plotting graph
plt.scatter(clusters['sepal_length'],clusters['sepal_width'],c=clusters['cluster_new_predict'],cmap='rainbow')
```

```python
#Wcss some of squares of each data point
wcss=[]
clim=10
#Clustering for n points
for i in range (1,clim):
    kmeans=KMeans(i)
    kmeans.fit(x_scaled)
    wcss_itr=kmeans.inertia_
    wcss.append(wcss_itr)
wcss
```

```python
#plotting graph
number_clusters = range(1,clim)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```
